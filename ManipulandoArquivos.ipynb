# Trabalhando com arquivos

```python
arquivo1 = open('./arquivo1.txt', 'r')
```

```python
type(arquivo1)
```

```python
print(arquivo1.read())
```

```python
#AQUI INFORMAMOS QUANTOS CARACTERES TEM NO ARQUIVO
print(arquivo1.tell())
```

###IMPORTANTE!!! - Quando lemos um arquivo, o cursor fica ao final de onde especificamos a leitura, ou seja, como eu li o arquivo inteiro o cursor esta no final do documento.

```python
#Se eu tentar le-lô novamente nada será impresso
print(arquivo1.read())
```

```python
#Para isso devemos por o cursor novamente para o inicio do documento usando o método seek
arquivo1.seek(0,0)
```

```python
#Como disse antes podemos especificar até onde queremos ler o documento, e obviamente o cursor irá parar no final de onde especificamos
print(arquivo1.read(23))
```

```python
# Repare que ele continuou a leitura de onde o cursor estava
print(arquivo1.read())
```

###Manipulando arquivos CSV

```python
csv = open('./salarios.csv','r')
```

```python
data = csv.read()
```

```python
rows = data.split('\n')
```

```python
array = []
for row in rows:
  split_row = row.split(',')
  array.append(split_row)
print(array)
```

```python
data_list = [row.split(',') for row in rows]
print(data_list)
```

```python
count = 0
for row in data_list:
  count += 1
print(count)
```

```python
columns = len(data_list[0])
print(columns)
```

```python
rows = len(data_list)
print(rows)
```

###Manipulando JSON

```python
from urllib.request import urlopen
```

```python
import json
```

```python
response = urlopen('https://vimeo.com/api/v2/video/57733101.json').read().decode('UTF8')
```

```python
data = json.loads(response)[0]
```

```python
for key, value in  data.items():
  print(key, value)
```

```python
with open('/content/contentJson.txt', 'w') as novo_arquivo:
  novo_arquivo.write(json.dumps(data))
```

```python
with open('/content/contentJson.txt', 'r') as arquivo_alvo:
  content_text = arquivo_alvo.read()
  with open('/content/contentCopyJson.txt', 'w') as arquivo_destino:
    arquivo_destino.write(content_text)

```

```python
#Abordagem simplificada
open('/content/contentCopy2Json.txt','w').write(open('/content/contentJson.txt','r').read())
```
